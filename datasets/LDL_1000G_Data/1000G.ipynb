### Extract Eur and AFR samples from 1000G

```python
# Import libraries
%pylab inline
import pandas as pd
import os
```

```python
sample = pd.read_csv('~/public/1000Genomes/igsr_samples.tsv',sep = '\t')
for anc in ['EUR','AFR']:
    sub_sample = sample[sample['Superpopulation code'] == anc][['Sample name']]
    sub_sample['IID'] = sub_sample['Sample name']
    sub_sample.to_csv(f'{anc}_sample.txt',sep = '\t',index = False,header = False)
    for i in range(1,23):
        os.system(f"plink --vcf ~/public/1000Genomes/ALL.chr{i}.phase3_shapeit2_mvncall_integrated_v5a.20130502.genotypes.vcf.gz \
                    --keep /home/hex002/ps2-wi24/{anc}_sample.txt \
                    --make-bed \
                    --out {anc}_{i}")
```
